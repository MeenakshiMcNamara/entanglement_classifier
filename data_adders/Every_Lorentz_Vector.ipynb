```python
import uproot
import matplotlib
import ipykernel
import numpy as np
import scipy
import torch
import pandas as pd
import sklearn
import uproot as uproot
import root_pandas
```

```python
depo_dir = '/depot/cms/top/mcnama20/TopSpinCorr-Run2-Entanglement/CMSSW_10_2_22/src/TopAnalysis/Configuration/analysis/diLeptonic/spinCorrInput_gendata_all/Nominal'
channel = "/mumu"  # can also be "/ee", /emu" and "/mumu"

file = channel + "_ttbarsignalplustau_fromDilepton_TuneCP5.root"

gen = True
```

```python
key = ""
if channel == "/ee":
    key = "ttBar_treeVariables_step8;21"
elif channel == "/emu":
    key = "ttBar_treeVariables_step8;88"
elif channel == "/mumu":
    key = "ttBar_treeVariables_step8;40"
else:
    print("Error!!!!")

prepend = ""
if gen:
    prepend = "gen_"
```

```python
data = uproot.open(depo_dir + channel + file, key = key)
data.keys()
```

```python
data = data[key]
data.keys()
```

```python
df = pd.DataFrame()
```

```python
# lepton
df[prepend + "l_eta"] = data['/' + prepend + 'l_eta'].array(library="np")
df[prepend + "lbar_eta"] = data['/' + prepend + 'lbar_eta'].array(library="np")

df[prepend + "l_phi"] = data['/' + prepend + 'l_phi'].array(library="np")
df[prepend +  "lbar_phi"] = data['/' + prepend + 'lbar_phi'].array(library="np")

df[prepend + "l_pt"] = data['/' + prepend + 'l_pt']. array(library="np")
df[prepend + "lbar_pt"] = data['/' + prepend + 'lbar_pt']. array(library="np")

df[prepend + "l_mass"] = data['/' + prepend + 'l_mass']. array(library="np")
df[prepend + "lbar_mass"] = data['/' + prepend + 'lbar_mass']. array(library="np")

# neutrino
df[prepend + "nu_eta"] = data['/' + prepend + 'nu_eta'].array(library="np")
df[prepend + "nubar_eta"] = data['/' + prepend + 'nubar_eta'].array(library="np")

df[prepend + "nu_phi"] = data['/' + prepend + 'nu_phi'].array(library="np")
df[prepend +  "nubar_phi"] = data['/' + prepend + 'nubar_phi'].array(library="np")

df[prepend + "nu_pt"] = data['/' + prepend + 'nu_pt']. array(library="np")
df[prepend + "nubar_pt"] = data['/' + prepend + 'nubar_pt']. array(library="np")

df[prepend + "nu_mass"] = data['/' + prepend + 'nu_mass']. array(library="np")
df[prepend + "nubar_mass"] = data['/' + prepend + 'nubar_mass']. array(library="np")

# b jet
df[prepend + "b_eta"] = data['/' + prepend + 'b_eta'].array(library="np")
df[prepend + "bbar_eta"] = data['/' + prepend + 'bbar_eta'].array(library="np")

df[prepend + "b_phi"] = data['/' + prepend + 'b_phi'].array(library="np")
df[prepend + "bbar_phi"] = data['/' + prepend + 'bbar_phi'].array(library="np")

df[prepend + "b_pt"] = data['/' + prepend + 'b_pt']. array(library="np")
df[prepend + "bbar_pt"] = data['/' + prepend + 'bbar_pt']. array(library="np")

df[prepend + "b_mass"] = data['/' + prepend + 'b_mass']. array(library="np")
df[prepend + "bbar_mass"] = data['/' + prepend + 'bbar_mass']. array(library="np")
# top quark
df[prepend + "top_eta"] = data['/' + prepend + 'top_eta'].array(library="np")
df[prepend + "tbar_eta"] = data['/' + prepend + 'tbar_eta'].array(library="np")

df[prepend + "top_phi"] = data['/' + prepend + 'top_phi'].array(library="np")
df[prepend + "tbar_phi"] = data['/' + prepend + 'tbar_phi'].array(library="np")

df[prepend + "top_pt"] = data['/' + prepend + 'top_pt']. array(library="np")
df[prepend + "tbar_pt"] = data['/' + prepend + 'tbar_pt']. array(library="np")

if not gen: 
    df[prepend + "top_mass"] = data['/' + prepend + 'top_mass']. array(library="np")
    df[prepend + "tbar_mass"] = data['/' + prepend + 'tbar_mass']. array(library="np")
else:
    df[prepend + "top_mass"] = 172.5
    df[prepend + "tbar_mass"] = 172.5
    
# wboson
df[prepend + "wplus_eta"] = data['/' + prepend + 'wplus_eta'].array(library="np")
df[prepend + "wminus_eta"] = data['/' + prepend + 'wminus_eta'].array(library="np")

df[prepend + "wplus_phi"] = data['/' + prepend + 'wplus_phi'].array(library="np")
df[prepend + "wminus_phi"] = data['/' + prepend + 'wminus_phi'].array(library="np")

df[prepend + "wplus_pt"] = data['/' + prepend + 'wplus_pt']. array(library="np")
df[prepend + "wminus_pt"] = data['/' + prepend + 'wminus_pt']. array(library="np")

df[prepend + "wminus_mass"] = data['/' + prepend + 'wplus_mass']. array(library="np")
df[prepend + "wplus_mass"] = data['/' + prepend + 'wminus_mass']. array(library="np")
```

```python
df["production_mode"] = data["/production_mode"].array(library = "np")
if not gen:
    df["eventWeight"] = data["/eventWeight"].array(library = "np")
else:
    df["eventWeight"] = data["/trueLevelWeight"].array(library = "np")
```
