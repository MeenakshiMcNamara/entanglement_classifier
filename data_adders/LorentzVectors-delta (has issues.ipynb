```python
import uproot
import matplotlib
import ipykernel
import numpy as np
import scipy
import torch
import pandas as pd
import sklearn
import uproot as uproot
import root_pandas
```

```python
depo_dir = '/depot/cms/top/mcnama20/TopSpinCorr-Run2-Entanglement/CMSSW_10_2_22/src/TopAnalysis/Configuration/analysis/diLeptonic/spinCorrInput_gendata_all/Nominal'
channel = "/emu"  # can also be "/ee", /emu" and "/mumu"

file = channel + "_ttbarsignalplustau_fromDilepton_TuneCP5.root"

gen = True

key = "ttBar_treeVariables_step8;88"  #can also be "ttBar_treeVariables_step8;21", "ttBar_treeVariables_step8;88" and "ttBar_treeVariables_step8;40"
```

```python
prepend = ""
if gen:
    prepend = "gen_"
```

```python
data = uproot.open(depo_dir + channel + file, key = key)
```

```python
data = data[key]
```

```python
df = pd.DataFrame()
```

```python
df["l_eta"] = data['/' + prepend + 'l_eta'].array(library="np")
df["lbar_eta"] = data['/' + prepend + 'lbar_eta'].array(library="np")

df["l_phi"] = data['/' + prepend + 'l_phi'].array(library="np")
df["lbar_phi"] = data['/' + prepend + 'lbar_phi'].array(library="np")

df["l_pt"] = data['/' + prepend + 'l_pt']. array(library="np")
df["lbar_pt"] = data['/' + prepend + 'lbar_pt']. array(library="np")

df["l_mass"] = data['/' + prepend + 'l_mass']. array(library="np")
df["lbar_mass"] = data['/' + prepend + 'lbar_mass']. array(library="np")

df["b_eta"] = data['/' + prepend + 'b_eta'].array(library="np")
df["bbar_eta"] = data['/' + prepend + 'bbar_eta'].array(library="np")

df["b_phi"] = data['/' + prepend + 'b_phi'].array(library="np")
df["bbar_phi"] = data['/' + prepend + 'bbar_phi'].array(library="np")

df["b_pt"] = data['/' + prepend + 'b_pt']. array(library="np")
df["bbar_pt"] = data['/' + prepend + 'bbar_pt']. array(library="np")

df["b_mass"] = data['/' + prepend + 'b_mass']. array(library="np")
df["bbar_mass"] = data['/' + prepend + 'bbar_mass']. array(library="np")
```

```python
df["llbar_delta_eta"] =  data['/' + prepend + 'llbar_delta_eta'].array(library="np")
df["bbbar_delta_eta"] =  data['/' + prepend + 'bbbar_delta_eta'].array(library="np")
df["lbbar_delta_eta"] =  data['/' + prepend + 'l_eta'].array(library="np") -  data['/bbar_eta'].array(library="np")
df["blbar_delta_eta"] =  data['/' + prepend + 'b_eta'].array(library="np") -  data['/lbar_eta'].array(library="np")
df["lb_delta_eta"] =  data['/' + prepend + 'l_eta'].array(library="np") -  data['/b_eta'].array(library="np")
df["bbarlbar_delta_eta"] =  data['/' + prepend + 'bbar_eta'].array(library="np") -  data['/lbar_eta'].array(library="np")

```

```python
df["llbar_delta_phi"] =  data['/' + prepend + 'llbar_delta_phi'].array(library="np")
df["bbbar_delta_phi"] =  data['/' + prepend + 'bbbar_delta_phi'].array(library="np")
df["lbbar_delta_phi"] =  data['/' + prepend + 'l_phi'].array(library="np") -  data['/bbar_phi'].array(library="np")
df["blbar_delta_phi"] =  data['/' + prepend + 'b_phi'].array(library="np") -  data['/lbar_phi'].array(library="np")
df["lb_delta_phi"] =  data['/' + prepend + 'l_phi'].array(library="np") -  data['/b_phi'].array(library="np")
df["bbarlbar_delta_phi"] =  data['/' + prepend + 'bbar_phi'].array(library="np") -  data['/lbar_phi'].array(library="np")

```

```python
df["production_mode"] = data["/production_mode"].array(library = "np")
if not gen:
    df["eventWeight"] = data["/eventWeight"].array(library = "np")
else:
    df["eventWeight"] = data["/trueLevelWeight"].array(library = "np")
```
