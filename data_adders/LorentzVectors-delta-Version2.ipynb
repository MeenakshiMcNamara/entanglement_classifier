```python
depo_dir = '/depot/cms/top/mcnama20/TopSpinCorr-Run2-Entanglement/CMSSW_10_2_22/src/TopAnalysis/Configuration/analysis/diLeptonic/spinCorrInput_wbosons_masses_all/Nominal'
channel = "/mumu"  # can also be "/ee", /emu" and "/mumu"

file = channel + "_ttbarsignalplustau_fromDilepton_TuneCP5.root"

key = "ttBar_treeVariables_step8;40"  #can also be "ttBar_treeVariables_step8;21", "ttBar_treeVariables_step8;88" and "ttBar_treeVariables_step8;40"
```

```python
df = pd.DataFrame()
```

```python
df["l_eta"] = data['/l_eta'].array(library="np")
df["lbar_eta"] = data['/lbar_eta'].array(library="np")

df["l_phi"] = data['/l_phi'].array(library="np")
df["lbar_phi"] = data['/lbar_phi'].array(library="np")

df["l_pt"] = data["/l_pt"]. array(library="np")
df["lbar_pt"] = data["/lbar_pt"]. array(library="np")

df["l_mass"] = data["/l_mass"]. array(library="np")
df["lbar_mass"] = data["lbar_mass"]. array(library="np")

df["b_eta"] = data['/b_eta'].array(library="np")
df["bbar_eta"] = data['/bbar_eta'].array(library="np")

df["b_phi"] = data['/b_phi'].array(library="np")
df["bbar_phi"] = data['/bbar_phi'].array(library="np")

df["b_pt"] = data["/b_pt"]. array(library="np")
df["bbar_pt"] = data["/bbar_pt"]. array(library="np")

df["b_mass"] = data["/b_mass"]. array(library="np")
df["bbar_mass"] = data["bbar_mass"]. array(library="np")
```

```python
df["llbar_delta_eta"] =  data['/llbar_delta_eta'].array(library="np")
df["bbbar_delta_eta"] =  data['/bbbar_delta_eta'].array(library="np")
df["lbbar_delta_eta"] =  data['/l_eta'].array(library="np") -  data['/bbar_eta'].array(library="np")
df["blbar_delta_eta"] =  data['/b_eta'].array(library="np") -  data['/lbar_eta'].array(library="np")
df["lb_delta_eta"] =  (data['/l_eta'].array(library="np") -  data['/b_eta'].array(library="np"))%np.pi
df["bbarlbar_delta_eta"] =  (data['/bbar_eta'].array(library="np") -  data['/lbar_eta'].array(library="np"))%np.pi

```

```python
df["llbar_delta_phi"] =  data['/llbar_delta_phi'].array(library="np")
df["bbbar_delta_phi"] =  data['/bbbar_delta_phi'].array(library="np")
df["lbbar_delta_phi"] = ( data['/l_phi'].array(library="np") -  data['/bbar_phi'].array(library="np"))%np.pi
df["blbar_delta_phi"] =  (data['/b_phi'].array(library="np") -  data['/lbar_phi'].array(library="np"))%np.pi
df["lb_delta_phi"] =  (data['/l_phi'].array(library="np") -  data['/b_phi'].array(library="np"))%np.pi
df["bbarlbar_delta_phi"] =  (data['/bbar_phi'].array(library="np") -  data['/lbar_phi'].array(library="np"))%np.pi

```

```python
df["production_mode"] = data["/production_mode"].array(library = "np")
df["eventWeight"] = data["/eventWeight"].array(library = "np")
```
